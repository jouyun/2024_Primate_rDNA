```python
import numpy as np
from cellpose import models
import tifffile
import glob
```

```python
model = models.CellposeModel(gpu=True, pretrained_model='../rDNA_models/models/v4')
```

```python
remake_chromosomes = True
if remake_chromosomes:
    for f in glob.glob('./rgb/*.tif'):
        img = tifffile.imread(f)
        img[:,:,1] = np.sqrt(img[:,:,1]).astype(np.uint8)

        chromosomes = model.eval([img], channels=[3,2], diameter=11,
                       normalize=True, resample=False)[0][0]
        img = tifffile.imread('./raw/'+f.split('\\')[-1])
        print(f)
        new_img = np.concatenate([img, [chromosomes]]).astype(np.single)
        tifffile.imwrite('./rDNA_inferred/'+f.split('\\')[-1], new_img
                                                                                    , imagej=True)
    
```
